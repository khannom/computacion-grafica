```python
def Comparativo(imagen1,imagen2):
    modificado=imagen1
    
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            if( modificado[i][j]>imagen2[i][j] ):
                modificado[i][j]=imagen2[i][j]

    return modificado

def Suma(imagen1,imagen2,a):
    modificado=imagen1
    
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            modificado[i][j]=imagen2[i][j]*a+modificado[i][j]*(1-a)

    return modificado

def Resta(imagen1,imagen2,a):
    modificado=imagen1
    
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            modificado[i][j]=modificado[i][j]-imagen2[i][j]*a
            #if (modificado[i][j]<imagen2[i][j]):
            #    modificado[i][j]=0
            #else:
            #    modificado[i][j]=modificado[i][j]-imagen2[i][j]
    return modificado

def Inversa(imagen):
    modificado=imagen
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            modificado[i][j]=255-imagen[i][j]
    return modificado

def Threshold(imagen):
    modificado=imagen
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            if(imagen[i][j] > 128):
                modificado[i][j] =255
            else:
                modificado[i][j] =0
    return modificado
```

```python
def andp(imagen1,imagen2):
    modificado=imagen1
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            modificado[i][j]=modificado[i][j]&imagen2[i][j]
    return modificado

def orp(imagen1,imagen2):
    modificado=imagen1
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            modificado[i][j]=modificado[i][j]|imagen2[i][j]
    return modificado

def xorp(imagen1,imagen2):
    modificado=imagen1
    for i in range(len(imagen1)):
        for j in range(len(imagen1[i])):
            modificado[i][j]=modificado[i][j]^imagen2[i][j]
    return modificado

def inversa(imagen):
    modificado=imagen
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            modificado[i][j]=255-imagen[i][j]
    return modificado

def Threshold(imagen,c):
    modificado=imagen
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            if(imagen[i][j] < c):
                modificado[i][j] =255
            else:
                modificado[i][j] =0
    return modificado
```
